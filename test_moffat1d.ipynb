```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.modeling.models import Moffat1D

fig = plt.figure()
ax = fig.add_subplot(111)
s1 = Moffat1D()
r = np.arange(-5, 5, .01)

for factor in np.arange(0, 2, 0.2):
    s1.amplitude = 5.0
    s1.gamma = 5.0
    s1.alpha = factor
    ax.plot(r, s1(r), lw=2, label='alpha = {}'.format(factor))

ax.legend()
plt.show()
```
